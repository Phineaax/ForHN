## 0.2 Variables 
##### D√©finition et logique
Une variable est une √©tiquette pour acc√©der √† un √©l√©ment de la m√©moire. Les variables sont modifiables √† l'infini. Attention, sur les jupyter notebook, l'affectation chronologique d'une variable ne d√©pend pas de l'ordre de pr√©sentation du notebook mais de l'ordre de lancement des cellules. √Ä l'int√©rieur des cellules en revanche, c'est bien dans l'ordre logique de haut en bas. 

##### Comportement selon le lieu de d√©finition
√âgalement, si vous d√©finissez une variable dans une condition, elle n'existe normalement qu'√† l'int√©rieur de la condition (si ce n'est pas le cas, inqui√©tez-vous). C'est important de d√©finir ses variables avant d'entrer dans une boucle ou une condition si on veut les utiliser en dehors. Sinon, √ßa risque de cr√©er des big probl√®mes. 

##### Comportement global
Pour les types simples(int, float, bool), l'√©tiquette appelle la valeur de la variable directement. Pour les autres types exotiques, typiquement les listes, l'√©tiquette appelle l'adresse du premier √©l√©ment de la liste. En mettant une liste dans une variable, on est donc directement face √† son adresse en m√©moire, et deux variables contenant la m√™me adresse de liste op√©reront l'une sur l'autre. 
En revanche, naturellement, si v1 = 3 et v2 = 3, si je fais v1 = v1+2, √ßa n'affectera pas v2. 

##### Tips
Enfin, 
* C'est important de mettre des noms parlants √† ses variables pour pouvoir relire son code plus tard (ce qui n'est pas √©vident du tout)
* Une variable ne prend canoniquement pas de majuscule (personne ne vous attaquera si vous le faites mais √ßa n'aide pas √† la lisibilit√©)

### 0.2 Blocs
Tout code se constitue en blocs imbriqu√©s les uns dans les autres. Ce sont des ensembles ordonn√©s d'instructions. En Python, chaque bloc est repr√©sent√© par l'indentation, et il est d√©pendant de toutes les conditions √† l'origine de son indentation. (Typiquement, s'il y a trois indentations avec un bloc, il faut que les conditions qui causent ces trois indentations soient respect√©es pour que le bloc soit run).

Une indentation (et donc un bloc) est cr√©e parce que :
* On suit une condition
* On entre dans une boucle
* On entre dans une fonction
Les instructions d'un bloc sont lanc√©es si toutes les conditions pour y arriver sont v√©rifi√©es, sinon on passe au bloc suivant dont les conditions sont respect√©es.

### 0.3 Conditions et Boucles
La base de la base : tout code est une succession de boucles et de conditions plus ou moins pimp√©es.


#### 0.3.1 Conditions
S√©parables en if / else / elif, ce sont des op√©rations qui prennent en argument (dans un sens) un bool√©en (valeur vrai ou faux). Ensuite, elles r√©alisent ou non une op√©ration en fonction de la valeur du bool√©en. 

Tr√®s concr√®tement : 
* if prend un bool√©en et r√©alise les actions associ√©es √† son bloc si il est True
* else ne fonctionne qu'apr√®s un if : le bloc else est r√©alis√© si la condition du if est False
* elif est l'√©quivalent d'un if dans un else : si la condition du premier if est fausse et que celle du deuxi√®me est vrai, on lance le bloc d'op√©rations.

Les conditions sont toujours de la forme : 
``` python
if (Condition √† valeur de bool√©en):
    bloc
elif (Condition √† valeur de bool√©en):
    bloc
else : # Pas de condition au else
    bloc
```

Les conditions ont pour op√©rateurs (liste approximativement exhaustive) : 
* ==
* \>, >=
* <, <=
* is (voir Tips 0.4)
* True / False
* N'importe quelle fonction renvoyant un bool√©en

Et pour coordinateurs :
* and
* or
* xor (rarement utilis√©, ou exclusif)

√âgalement, l'op√©rateur **not** permet de passer toute la formule au n√©gatif.

````python
# En exemple:
print("Premier exemple : ")
if ("A" in "Ad√®le"):
    print("Il y a un A dans Ad√®le")
elif (len("Ad√®le") == 5):
    print("Il y a 5 lettres dans Ad√®le")
else :
    print("Je n'ai rien √† dire")

# Ici on print la condition du if qui est remplie, mais pas les deux autres parce qu'elles d√©pendent du fait que le if ne soit pas rempli. 
# Ils seraient rempli s'il n'y avait pas de A dans Ad√®le.

# Par ailleurs, ("A" in "Ad√®le") et (len("Ad√®le") == 5) sont des bool√©ens √©gaux √† True. 

print("\nDeuxi√®me exemple : ")
if not ("A" in "Ad√®le" and "B" in "Bernard"):  # Ici les deux sont vrais donc la condition entre parenth√®ses est vraie, mais le not la renverse
    print("Mince")                             # Donc le global est faux.
elif ("A" in "Albert" or "B" in "Tabarnak"):   # Ici, la premi√®re condition est vraie, donc comme c'est un ou inclusif, il n'a m√™me pas besoin
    print("Ahh ! Je me disais bien!")          # de regarder la deuxi√®me variable pour savoir que c'est vrai 
                                               # D'ailleurs, dans les faits, il ne le fait pas, c'est l'execution paresseuse : 
# Pour and : si le premier √©l√©ment est faux, il ne regarde pas le deuxi√®me
# Pour ou, si le premier est vrai, il ne regarde pas le deuxi√®me. 

print("\nTroisi√®me exemple : ")
if not "A" in "Ad√®le" or "B" in "Bernard":   # Attention, ici comme il n'y a pas de parenth√®ses, la premi√®re partie est rendue fausse mais la 
    print("Je le savais!")                   # deuxi√®me reste vraie.
else :  
    print("Hm")
````

#### 0.3.2 Boucles
Les boucles sont des blocs sous une condition qui se r√©p√®tent tant que la condition est vraie. 

Il y a deux types de boucles : 
* for : s'incr√©mente seule
* while : √† incr√©menter √† la main

Les conditions sont toujours de la forme : 
``` python
for var in structure: # Avec var : une variable it√©rable √† l'int√©rieur structure : une liste / un dictionnaire / une s√©rie...
    bloc              # var peut aussi s'adpter en (var1, var2,...) sous forme d'un tuple si structure contient des tuples.

while (Condition √† valeur de bool√©en):
    bloc
    -> Action qui peut modifier la valeur de la condition !

```

Elles servent en diff√©rentes occasions, mais on utilise quasiment exclusivement le for √† haut niveau. On ne peut utiliser que la boucle for dans les compr√©hensions, et de mani√®re g√©n√©rale, elle est plus compacte et √©vite les erreurs cr√©ant des boucles infinies.

---------------------------------------------------------------------------------------------------------------------------------------------
Il est possible, mais tr√®s peu recommand√©, d'utiliser les mots clefs **pass** et **break** dans les boucles et conditions : 
* pass zappe la fin du bloc et passe √† la prochaine it√©ration
* break brise la boucle : ne finit pas le bloc et passe aux instructions apr√®s la boucle
  
C'est peu recommand√© de les utiliser parce qu'il y a quasiment toujours un moyen plus propre de finir son bloc. 

````python
# En exemple:
# Boucle for:
for i in range(10): # Parcourt automatiquement chaque √©l√©ment de range(10), qui est une liste des 10 premiers entiers. 
    #print(i)
    pass  # Ici, pass me sert uniquement √† √©viter une erreur due au fait que mon bloc est vide quand je commente print.

# Boucle while:
i = 0
while (i < 10): # Ici la condition est plus claire, mais on doit nous-m√™me faire augmenter i. Il y a le risque si on oublie d'avoir une boucle
    #print(i)   # infinie, ce qui est un enfer. Donc en pratique, sauf quand for ne peut pas faire ce qu'on veut, on n'utilise pas while.
    i+=1


# Boucles imbriqu√©e : 
# Quand une boucle est dans une autre, logiquement (mais c'est utile de le conscientiser), elle r√©alise toutes les op√©rations de la boucle 
# int√©rieure avant de passer √† la prochaine it√©ration de la boucle ext√©rieure. Alors, on recommence la boucle int√©rieure et ainsi de suite.
l = list()
for i in range(5):
    for j in range(5):
        l.append((i, j))
        
#print(l) # On voit que d'abord on traite toutes les valeurs de j pour un m√™me i, et ensuite on change la valeur de i.
````

### 0.3 Fonctions
Les fonctions sont tr√®s simplement des blocs r√©sum√©s en une √©tiquette. Elles prennent des arguments qui permettent de donner les valeurs que l'on veut aux variables utilis√©es dans le bloc fonction. Elles ont √©galement une valeur de retour. 

Selon les fonctions, elles peuvent modifier nativement l'objet donn√© en argument (typiquement en modifiant une liste), ou bien renvoyer un objet voulu. Elles peuvent √©galement faire les deux

Les fonctions sont toujours de la forme :
```python
def nomFonction(arg1, arg2...):
    bloc
    return var # Qui peut se d√©cliner en tous types de variables, y compris des tuples si besoin de retourner plusieurs valeurs
```

````python
# En exemple : 
print("Fonction 1 : ")
def fonction(gentil, mechant):
    if (gentil > mechant):
        print("Trop chouette !")
        return "WIN"
    else : 
        print("Sapristi !")
        return "LOSE"

val = fonction(2, 1)
print("Retour : ", val)

print("\nFonction 2 : ")
def fonctionDuDiable(l):
    l2 = l.copy()  # Permet de copier une liste dans une autre
    l.append("Mouahahaha nouvelle valeur inattendue")
    return l2

l = list(range(10))
l2 = fonctionDuDiable(l)
print("l : ", l, "l2 : ", l2)
````

Comme les listes, les dictionnaires sont des objets muables et dynamiques. Ils peuvent √™tre modifi√©s et s'√©tendre selon vos besoins.
Un dictionnaire peut contenir des objets de n'importe quel type et m√™me inclure d'autres dictionnaires.

## 5. Tuples

### 5.1 Concept
Les tuples en Python sont des collections ordonn√©es et immuables de valeurs. Ils sont similaires aux listes, mais les tuples sont immuables, ce qui signifie que leurs √©l√©ments ne peuvent pas √™tre modifi√©s une fois qu'ils ont √©t√© d√©finis. (Merci internet *https://blog.stephane-robert.info*)

### 5.3 Utilisation
Pas mal de fonctions de base renvoient des tuples, c'est utile de savoir les manipuler. √áa permet √©galement de faire renvoyer plusieurs valeurs √† une fonction.

````python
lieu = ["Maison", "Appartement", "Sous-sol", "Bunker", "Cave sombre"]
date = [1897, 1950, 1550, 1940, 380]
proprietaire = ["Glauque", "Proprio", "Gentil", "M√©chant", "Inconnu"]

dico = dict()
for l, d, p in zip(lieu, date, proprietaire):    # zip est une fonction qui cr√©e des tuples correspondant aux √©l√©ments de chaque liste donn√©e
    if (l in dico):                              # en argument mis √† la suite les uns en face des autres (je sais pas si c'est clair, sinon
        dico[l].append(d, p)                     # regardez juste le print). C'est sous forme d'un format zip qui est une esp√®ce de liste de tuples
    else : 
        dico[l] = [(d, p)]

#print(dico)
for (k, v) in dico.items():
    (d, p) = v[0]
    print(p, "est propri√©taire d'un.e", k, "datant de", d)
````
